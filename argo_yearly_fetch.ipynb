```python
from argopy import DataFetcher as ArgoData
import pandas as pd
import os
```

YEAR- 2023


```python

regions = {
    "Arabian Sea": [65, 75, 12, 20, 0, 500],
    "Bay of Bengal": [80, 95, 10, 20, 0, 500],
    "Equatorial IO": [60, 90, -5, 5, 0, 500],
    "Southern IO": [50, 70, -35, -25, 0, 500],
}


months_2023 = [
    ("2023-01-01", "2023-02-01"),
    ("2023-02-01", "2023-03-01"),
    ("2023-03-01", "2023-04-01"),
    ("2023-04-01", "2023-05-01"),
    ("2023-05-01", "2023-06-01"),
    ("2023-06-01", "2023-07-01"),
    ("2023-07-01", "2023-08-01"),
    ("2023-08-01", "2023-09-01"),
    ("2023-09-01", "2023-10-01"),
    ("2023-10-01", "2023-11-01"),
    ("2023-11-01", "2023-12-01"),
    ("2023-12-01", "2024-01-01"),
]

output_csv = "argo_2023_multiregion.csv"
first_write = True


os.makedirs("netcdf_files", exist_ok=True)


for region_name, region_coords in regions.items():
    print(f"\n Region: {region_name}")

    for start, end in months_2023:
        print(f"   {start} → {end}")

        try:
            full_region = region_coords + [start, end]


            argo = ArgoData().region(full_region)
            ds = argo.to_xarray()

            safe_region_name = region_name.replace(" ", "_")
            nc_filename = f"netcdf_files/{safe_region_name}_{start}.nc"
            ds.to_netcdf(nc_filename)

            df = ds.to_dataframe().reset_index()

            df["region"] = region_name
            df["year"] = 2023

            df.to_csv(
                output_csv,
                mode='a',
                header=first_write,
                index=False
            )

            first_write = False

        except Exception as e:
            print(f"   Skipped: {e}")

print("\n Data collection for 2023 complete!")
```

```python
import xarray as xr

folder = "netcdf_files2023"

datasets = []

for file in os.listdir(folder):
    if file.endswith(".nc"):
        path = os.path.join(folder, file)
        ds = xr.open_dataset(path)
        datasets.append(ds)
```

```python
combined_ds = xr.concat(datasets, dim="N_POINTS")
```

```python
combined_ds.to_netcdf("argo_2023_full.nc")
```
